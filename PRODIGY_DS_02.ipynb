```python
#Exploratory data analysis (EDA) to explore relationships between variables and identify patterns and trends in the data.
#Using seaborn and matplotlib libraries to create visualizations.

# Visualize survival count by gender
sns.countplot(x='Survived', hue='Sex', data=titanic_df, palette=['skyblue', 'darkblue'])
plt.title('Survival Count by Gender')
plt.show()
```

```python
# Visualize survival count by passenger class
sns.countplot(x='Survived', hue='Pclass', data=titanic_df, palette = ['turquoise', 'orange', 'purple'])
plt.title('Survival Count by Passenger Class')
plt.show()
```

```python
# Visualize age distribution of passengers
sns.histplot(data=titanic_df, x='Age', bins=20, kde=True, color = 'gold')
plt.title('Age Distribution of Passengers')
plt.xlabel('Age')
plt.ylabel('Count')
plt.show()
```

```python
# Select only numeric columns for correlation analysis
numeric_columns = titan_df.select_dtypes(include=['int64', 'float64'])

# Calculate correlation matrix
correlation_matrix = numeric_columns.corr()
```

```python
# Plot correlation matrix as a heatmap
plt.figure(figsize=(10, 8))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt=".2f")
plt.title('Correlation Matrix')
plt.show()
```
